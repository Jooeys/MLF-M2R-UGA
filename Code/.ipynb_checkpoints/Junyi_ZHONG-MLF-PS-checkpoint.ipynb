## Step 2. 
Code the Perceptron, Adaline and LR (Logistic regression) programs

Hint: You can write a function that calculates the dot product between an example $\mathbf{x} = (x_1, \ldots, x_d)$ and the weight vector $\mathbf{w} = (w_0, w_1, \ldots, w_d)$: 
$ h(\mathbf{x},\mathbf{w}) = w_0 + \ sum_ {j = 1} ^ d w_j x_j $.

x: input vector,characteristic vector

```python
from math import exp
from operator import add

def h(x,w):
    # The prediction of the model
    # Pred=0.0
    Pred = w[0]
    # Calculates the dot product
    for i in range(len(x)):
        Pred += w[i + 1] * x[i]
    return Pred

def Perceptron(Train,eta,MaxEp):
    # Perceptron Algorithm
    epoch=0
    err=1
    E=[]
    
    d=len(Train[0])-1
    m=len(Train)
    
    W=[0.0 for i in range(d+1)]
    
    while (epoch < MaxEp and err !=0 ):
        err = 0
        for x in Train:
            dotproduct = h(x[0:d],W)
            if(x[d] * dotproduct <= 0):
                err += 1
                W[0] += eta * x[d]
                for i in range(d):
                    W[i+1] += eta * x[d] * x[i]
        E.append(err / len(Train))
        epoch += 1
    return W,err

def Adaline(Train,eta,MaxEp,d):
    # Adaline Algorithm 
    epoch=0
    err=0
    E=[]
    
    d=len(Train[0])-1
    m=len(Train)
    W=[0.0 for i in range(d+1)]
    
    for x in Train:
        W[0] += 2 * eta * (x[d] - h(x,W))
        for j in range(d):
            W[j + 1] += 2 * eta * (x[d] - h(x,W))
        if(x[d] * h(x,W) <=0):
            err += 1
    return W,err

def Adaline(Train,W,eta,d):
    err=0
    for x in Train:
        W[0]+=2*eta*(x[d]-h(x,W))
        for j in range(d):
            W[j+1]+=2*eta*x[j]*(x[d]-h(x,W))
        if(x[d]*h(x,W)<=0):
            err+=1
    return W,err

def sigma(z):
    return (1/(1+exp(-z))
            
def LR(Train,W,eta,d):
    # Logisitc Regression Algorithm 
    epoch=0
    err=1
    E=[]
    d=len(Train[0])-1
    m=len(Train)
    W=[0.0 for i in range(d+1)]
    
    for x in Train:
        errlog = (1.0 - sigma(x[d] * h(x,W)))
        W[0] += eta * (x[d] * errlog)
        for j in range(d):
            W[j+1] += eta * (x[d] * x[j] * errlog)
        if(x[d] * h(x,W) <= 0):
            err += 1
    
    return W,err
```

## Step 4. Read training set
We are now going to focus on the behavior of the three models on http://archive.ics.uci.edu/ml/datasets/connectionist+bench+(sonar,+mines+vs.+rocks), https://archive.ics.uci.edu/ml/datasets/spambase, https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Original%29, https://archive.ics.uci.edu/ml/datasets/Ionosphere. These files are in the current respository with the names `sonar.txt`; `spam.txt`; `wdbc.txt` and `ionoshpere.txt`. We can use the following `ReadCollection` function in order to read the files in the form of the training set that is requested. 

```python
from math import sqrt
import pandas as pd
import random
from sklearn.model_selection import train_test_split

def Normalize(x):
    norm=0.0
    for e in x:
        norm+=e**2
    for i in range(len(x)):
        x[i]/=sqrt(norm)
    return x

# Read wdbc.txt file in the Python format of request training set 
def ReadCollection(filename):
    tag_df=pd.read_table(filename,sep=',',header=None)
    Dic={'M': -1, 'B': +1}
    X=[]
    for e in range(len(tag_df)):
        x=list(tag_df.loc[e,:])
        x.pop(0)
        cls=x.pop(0)
        x=Normalize(x)
        x.insert(len(x),Dic[cls])
        X.append(x)

    
    random.shuffle(X)

    return X
```

```python
def EmpiricalRisk(Test,W):
    E=0.0
    m=len(Test)
    
    # The empirical error of a model with weight W on a test set of size m
    
    return E/float(m)

X=ReadCollection("/Users/junyi/Desktop/00UGA-M2/MLF-M2R-UGA/Data/wdbc.txt")

errP=errA=errL=0.0

# set learning rate 0.1
for i in range(20):
    x_train ,x_test = train_test_split(X,test_size=0.25)
    WLP=Perceptron(x_train,0.1,500)
    errP+=EmpiricalRisk(x_test,WLP)
    
    WLA=Adaline(x_train,0.1,500)
    errA+=EmpiricalRisk(x_test,WLA)
    
    WLR=LR(x_train,0.1,500)
    errL+=EmpiricalRisk(x_test,WLR)
    
print("Err perceptron=\n",errP/float(20),"Err Adaline=",errA/float(20),"Err RL=",errL/float(20))
print("--------------------------------------------")

# set learning rate 0.01
for i in range(20):
    x_train ,x_test = train_test_split(X,test_size=0.25)
    WLP=Perceptron(x_train,0.01,500)
    errP+=EmpiricalRisk(x_test,WLP)
    WLA=Adaline(x_train,0.01,500)
    errA+=EmpiricalRisk(x_test,WLA)
    WLR=LR(x_train,0.01,500)
    errL+=EmpiricalRisk(x_test,WLR)
    
print("Err perceptron=\n",errP/float(20),"Err Adaline=",errA/float(20),"Err RL=",errL/float(20))
```
